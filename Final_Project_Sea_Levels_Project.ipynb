### Exercise 4 - plot line of best fit for CSIRO

*   Use the `linregress` function from `scipi.stats` to get the slope and y-intercept of the line of best fit.
*   Plot the line of best fit over the top of the scatter plot.
*   Make the line go through the year 2050 to predict the sea level rise in 2050.    

*(Hint:  use numpy to create an array from the first year to 2050 with an increment of 1)*

**CORRELATION WITH DATA**

**In this exercise, I have to plot line of best fit for CSIRO.For this I have to find correlation between variables. I will import linregress from scipy.stats to find linear regression.**

**SciPy is a scientific computation library that uses NumPy underneath.**
**SciPy stands for Scientific Python.**

**Linear regression uses the relationship between the data-points to draw a straight line through all them.**

**This line can be used to predict future values.**

**I will find slope and intercept values those will be used in drawing the best fit line over the top of the scatter plot.**

**For creating the scatter plot, I will use the same method as I used in the earlier excercises in this project.I will use the dataframe "Data" to plot a scatter plot with assigning the variables to axises.**


**To decide the range on x axis for generating the sequence of number of years for how many years I want to predict the result, I will use numpy as np. So I will import numpy also.**

**NumPy is a Python library.**

**NumPy is used for working with arrays.**

**NumPy is short for "Numerical Python".**

```python
from scipy.stats import linregress
import numpy as np

x=data['Year']    # data is the name of dataframe
y=data['CSIRO Adjusted Sea Level']


# get slope and y-intercept
regression=linregress(x,y)
#print(regression)             # if you want to print regression, slope, intercept , then remove # sign before them.


slope=regression.slope
#print(slope)

intercept=regression.intercept
#print("Intercept is:",intercept)



# to get the scatter plot
data.plot(kind='scatter', x= 'Year', y= 'CSIRO Adjusted Sea Level')


# create an extended range for x to predict sea level in 2050

# FOR THIS-
# Create two new list to assign number -
# x= 0 to 2050 for year( how many years are asked) and y = slope*x+intercept where slope is m and intercept is c in the formula y=mx+c where y is the new value dependent on the values on x axis.


#Create a range of years from the start year to 2050

x=np.arange(x.min(),2050)              # arange() is a NumPy function used to generate numerical sequences.
                                       #arange() generates a sequence of values from start to stop with a given step size
y=slope*x+intercept  # y=mx+c


# draw the scatter plot with line of best fit
plt.plot(x,y)



```

**I have generated the regression and found the slope and intercept values, created the scatter plot, defined the range for years from minimum to 2050 to predict the sea level rise in 2050 in the form of plotting best fit line. To plot best fit line , I used the linear equation y=mx+c that shows the result depending upon the values on x axis using the slope and intercept to y.**

**This best fit line shows the constant increment in sea level in 2050.**

### Exercise 5 - plot line of best fit for NOAA
---

*   Use the `linregress` function from `scipi.stats` to get the slope and y-intercept of the line of best fit.
*   Plot the line of best fit over the top of the scatter plot.
*   Make the line go through the year 2030 to predict the sea level rise in 2030

*(Hint:  use numpy to create an array from the first year to 2030 with an increment of 1 AND remember to use noaa, which has NaN rows removed)*

**CORRELATION AND REGRESSION ANALYSIS**

**This is the similar exercise as above. I will follow the same method. I will  only change the dataframe(that is cleaned with no null values in NOAA column) and will set the range of number of years to 2030 to get best fit line for NOAA in 2030**

```python
# get scope and y-intercept

from scipy.stats import linregress
import numpy as np

x=df_new['Year']     # here our cleaned dataframe is new_df that we are using .
y=df_new['NOAA Adjusted Sea Level']


# get slope and y-intercept
regression=linregress(x,y)
#print(regression)          # remove # sign in front of print regression,slope and intercept if you want to print


slope=regression.slope
#print(slope)

intercept=regression.intercept
#print("Intercept is:",intercept)

# to draw a scatter plot first

# we can do it in two ways-
#1-
df_new.plot(kind="scatter", x="Year" , y="NOAA Adjusted Sea Level")

#2-
plt.scatter(x,y)

# create an extended range for x to predict sea level in 2030

x=np.arange(x.min(),2030)
y=slope*x+intercept

# draw the scatter plot with line of best fit

plt.plot(x,y)

```


### Exercise 6 - shorter range (starting 2000)
---

Plot a new line of best fit just using the data from year 2000 through the most recent year in the dataset and using the CSIRO adjusted levels. Make the line also go through the year 2050 to predict the sea level rise in 2050 if the rate of rise continues as it has since the year 2000.

**I will be doing this exercise as above. Just need to create a new variable for the column "Year" that will start from year 2000. Rest of the process will be same.**

```python
# get scope and y-intercept

from scipy.stats import linregress
import numpy as np

data=data[data['Year']>=2000]
x=data['Year']
y=data["CSIRO Adjusted Sea Level"]

regression=linregress(x,y)


slope=regression.slope
intercept=regression.intercept

data.plot(kind="scatter", x="Year",y="CSIRO Adjusted Sea Level")



# create an extended range for x to predict sea level in 2050

x=np.arange(x.min(),2050)
y=slope*x+intercept


# draw the scatter plot with line of best fit

plt.plot(x,y)
```

### Exercise 6 - add labels to the axes  
---

The x label should be "Year", the y label should be "Sea Level (inches)", and the title should be "Rise in Sea Level"

**I will be doing same steps to draw a scatter plot and line of best fit for CSIRO from the year 2000 to 2050 then will label the axises and title.**

**With Pyplot, we can use the xlabel() and ylabel() functions to set a label for the x- and y-axis. For this just need to add lablels to the axis.**

**With Pyplot, we can use the title() function to set a title for the plot. We need to add the plot title.**

```python
# get scope and y-intercept

from scipy.stats import linregress
import numpy as np

import matplotlib.pyplot as plt


data=data[data['Year']>=2000]
x=data['Year']
y=data["CSIRO Adjusted Sea Level"]

regression=linregress(x,y)


slope=regression.slope
intercept=regression.intercept

data.plot(kind="scatter", x="Year",y="CSIRO Adjusted Sea Level")



# create an extended range for x to predict sea level in 2050

x=np.arange(x.min(),2050)
y=slope*x+intercept


# draw the scatter plot with line of best fit

plt.plot(x,y)


#Adding the labels to the axises using pyplot
# The x label should be "Year", the y label should be "Sea Level (inches)", and the title should be "Rise in Sea Level"

plt.xlabel("Year")
plt.ylabel("Sea Level (inches)")
plt.title("Rise in Sea Level")

plt.plot(x,y)
```
