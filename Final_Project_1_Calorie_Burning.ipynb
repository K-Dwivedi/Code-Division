**Cleaning Empty cells first**

**I will be dealing with empty cells here. Although there are many ways to remove empty cell or fill with any calculated value but I think I should go with mode because mode is the most occuring value of the dataset so I can fill the empty cell with mode and probably it won't affect adversly to my dataset.**

```python
print(df)  # get the dataframe

# there are 2 NaN values means empty cells in Calorie column. We can use mode(most occured value) to fill the empty place.

x=df["Calories"].mode()[0]             # function to get the mode of the 'Calories' column
df["Calories"].fillna(x, inplace = True)
print(df.to_string())

# In the Calories column and row number 18 and 28 there were NaN value that is now filled with mode of the Calories series that is 300.



```

**Formatting:**

**Converting the data in the correct format**

**Data formatting is the process of transforming data into a common format, which helps us to perform comparisons.**

**In this dataset, there are two records of wrong format in 'Date'column in row 22 and 26.**

**'Date' column should be a string that represents a date:**  
**I have to convert all the cells in 'Date' column into same format.**

**Converting all cells of 'Date' column in dates**

```python
df['Date'] = pd.to_datetime(df['Date'])           # function to format all the cells of the 'Date column into date format.

print(df.to_string())      # now row 26 has been converted in date format but row 22 is showing Not a Time value that is a empty value.
                           # I will remove this entire row 22 in my next step.
```

**Removing rows**



**As row 22 in 'Date' column is an empty cell so I will remove this row so that my dataset can be in required format.**

```python
# to remove a row with null value , we use dropna() function

    # the two lines of importing pandas and reading the file path through pandas,
      # and defining the dataframe is not necessary to write again and again if we are using the same dataset or file.



df
df['Date'] = pd.to_datetime(df['Date'])

df.dropna(subset=['Date'], inplace = True)

print(df.to_string())            # as we can see the row 22 is removed now.

```

**As we can see the row 22 is removed from the dataframe now.**


**Replacing Values**

 **As I can see that in row 7, the duration is 450, but for all the other rows the duration is between 30 and 60.**


**So it is a wrong data as its not possible for anyone to workout 450 min according to this dataset.**


**# We can replace the value or can delete the row or can use loop to restrict the wrong values.**

**I assume that its a typo error that is mistyped as 450 in place of 45 so I will change it in 45.**

```python
df.loc[7, "Duration"] = 45  # let's suppose the value should be "45" instead of "450" in column "Duration" row 7, and we could just insert "45" in row 7
print(df.to_string())                # now we can see 45 in place of 450 in row 7 column "Duration".
```

**Removing Duplicates:**

**Duplicate rows are rows that have been registered more than one time.**

```python
# we can see that in rows 11 and 12 there are same values means these are duplicate values.
# to find the duplicates we use the function below-

df.duplicated()         # in row 12 we can see True (means it is a duplicate value), False means there is no duplicate in that row.
                      # we need to remove the duplicate row
```

**To remove the duplicate row**

```python
df.drop_duplicates(inplace=True)                 # function to remove duplicate value
print(df.to_string())          # now the row 12 has been removed from the result
```

**Analyzing the relation among columns**

**Now I am taking only two columns 'Duration' and 'Calories' to find the relationship between them as I think these two columns are useful to tell about the workout results.**

**STATISTICS**

**To get the min,max,std,mean (average) etc.of-"Duration" of workout and "Calories" burnt in that duration:**

**As I want to see the minimum,maximum and average of Duration of workout  and its corresponding effect on Calories burnt so I would go with the aggregate of statistics for both columns.**

**Thus I will get the data about how much duration of time is required for burning the approx count of calories.**

```python
df_new.agg(                                           # function To display min,max and mean value of time duration of workout and calories burnt.
    { "Duration": ["min", "max", "mean"],
         "Calories": ["min", "max", "mean"]
      }
)
```
